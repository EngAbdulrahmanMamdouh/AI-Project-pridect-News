```python
import pandas as pd
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
nltk.download('punkt') 
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
def remove_noise(text):
    text = re.sub(r'@\w+', '', text)                   # remove mentions (@user)
    text = re.sub(r'#\w+', '', text)                   # remove hashtags (#word)
    text = re.sub(r'http\S+|www\S+', '', text)         # remove links
    text = re.sub(r'\d+', '', text)                    # remove numbers
    text = re.sub(r'[^\x00-\x7F\u0600-\u06FF]', '', text)  # remove emojis & special chars
    return text
```

```python
def to_lowercase(text):
    return text.lower()
```

```python
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))
```

```python
def tokenize_text(text):
    return word_tokenize(text)
```

```python
def remove_stopwords(tokens):
    return [t for t in tokens if t not in stop_words]
```

```python
def apply_stemming(tokens):
    return [stemmer.stem(t) for t in tokens]
```

```python
def apply_lemmatization(tokens):
    return [lemmatizer.lemmatize(t) for t in tokens]
```

```python
def preprocess_text(text):
    if not isinstance(text, str):
        return ""

    text = remove_noise(text)
    text = to_lowercase(text)
    text = remove_punctuation(text)
    tokens = tokenize_text(text)
    tokens = remove_stopwords(tokens)
    tokens = apply_stemming(tokens)
    tokens = apply_lemmatization(tokens)

    return " ".join(tokens)
```

```python
news['clean_short_description'] = news['short_description'].apply(preprocess_text)
```

```python
news['clean_headline'] = news['headline'].apply(preprocess_text)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
# دمج العنوان والوصف في عمود جديد
news['text'] = news['headline'] + " " + news['short_description']

# بعد كده 'text' ده هيكون X
X = news['text']

# و Y هو العمود اللي عايز تتنبأ بيه
y = news['category']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer  #

tfidf = TfidfVectorizer(
    stop_words='english',
    max_features=15000                      #اقصي عدد كلمات بيستخدمها الموديل عشان التدريب الكلمات الغير شائعة
)

X_tfidf = tfidf.fit_transform(X)

X_tfidf
```

```python
#تقسيم الداتا الي # train test split (80% train - 20% test)
X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf, y, test_size=0.2, random_state=42
)
print('Training Data :', X_train.shape)
print('Testing Data : ', X_test.shape)
print('testing Data :', y_train.shape)
print('testing Data : ', y_test.shape)
```

```python
model = LogisticRegression(
    max_iter=1000,
    n_jobs=-1
)

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

y_pred
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Precision
precision = precision_score(y_test, y_pred, average='weighted')
print("Precision:", precision)

# Recall
recall = recall_score(y_test, y_pred, average='weighted')
print("Recall:", recall)

# F1-score
f1 = f1_score(y_test, y_pred, average='weighted')
print("F1-score:", f1)

# Report كامل
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

```python
import joblib

joblib.dump(model, 'logistic_model.pkl')
print("Successfully")
```

```python
# validation data
X_test_sample = news[['headline', 'short_description', 'category']] \
                    .sample(n=5000, random_state=42)

X_test_text = X_test_sample['headline'] + " " + X_test_sample['short_description']
y_true = X_test_sample['category']

X_test_vec = tfidf.transform(X_test_text)

y_pred = model.predict(X_test_vec)

accuracy = accuracy_score(y_true, y_pred)
print("Accuracy:", accuracy * 100, "%")
```
